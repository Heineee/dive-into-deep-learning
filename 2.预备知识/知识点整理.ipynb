# 《动手学深度学习》第2章 预备知识 知识点整理
## 2.1 数据操作
- **入门**：张量（多维数组）是深度学习数据的基本载体，类似NumPy数组，支持GPU加速、自动微分。
- **运算符**：支持加减乘除、幂、取模、比较等逐元素运算。
- **广播机制**：不同形状张量运算时，自动扩展维度匹配形状，无需复制数据。
- **索引和切片**：按位置、范围取元素，支持修改张量内容。
- **节省内存**：使用原地操作（如 `+=`）避免创建新张量，减少内存占用。
- **转换为其他Python对象**：张量与NumPy数组、Python列表相互转换。

## 2.2 数据预处理
- **读取数据集**：加载CSV等格式数据，表格数据行=样本、列=特征。
- **处理缺失值**：
  - 数值特征：均值/中位数填充；
  - 类别特征：最频繁值填充或新增“缺失”类别。
- **转换为张量格式**：将预处理后数据转为张量，适配模型输入。

## 2.3 线性代数
- **标量**：单个数值，0维张量。
- **向量**：一维数组，1维张量，代表特征/样本。
- **矩阵**：二维数组，2维张量，样本-特征矩阵常用。
- **张量**：多维数组（≥3维），深度学习数据主流形式。
- **张量算法基本性质**：交换律、结合律、分配律。
- **降维**：求和、求均值、最大值等聚合操作，降低张量维度。
- **点积**：两向量对应元素相乘求和，结果为标量。
- **矩阵-向量积**：矩阵与向量相乘，输出新向量。
- **矩阵-矩阵乘法**：行乘列求和，结果为新矩阵。
- **范数**：衡量张量大小，常用L1范数、L2范数。

## 2.4 微积分
- **导数和微分**：导数描述函数瞬时变化率，微分是微小变化量。
- **偏导数**：多元函数对单个变量的导数，固定其他变量。
- **梯度**：多元函数所有偏导数组成的向量，指向函数上升最快方向。
- **链式法则**：复合函数求导规则，深度学习反向传播核心。

## 2.5 自动微分
- **简单例子**：定义函数、计算梯度，无需手动推导导数公式。
- **非标量变量反向传播**：支持张量输出的梯度计算。
- **分离计算**：`detach()` 中断梯度流，冻结部分参数。
- **Python控制流梯度计算**：支持条件、循环语句的自动微分。

## 2.6 概率
- **基本概率论**：事件、概率、条件概率、独立性。
- **处理多个随机变量**：联合概率、边缘概率、条件概率、贝叶斯定理。
- **期望和方差**：期望=随机变量均值，方差=偏离期望程度，衡量波动。

## 2.7 查阅文档
- **查找模块中的所有函数和类**：快速定位模块功能。
- **查找特定函数和类的用法**：查询参数、返回值、示例，解决实操问题。
