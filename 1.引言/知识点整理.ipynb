# 《动手学深度学习》第1章 引言 知识点整理
## 1.1 人工智能、机器学习与深度学习
### 1.1.1 传统编程 vs 机器学习
- **传统编程**：人工定义规则与逻辑，程序逻辑固定，无法自动优化
- **机器学习**：从**数据经验**中学习规律，无需手动编写完整规则，性能随数据迭代提升，本质是**用数据编程**

### 1.1.2 三者层级关系
1. **人工智能(AI)**：让机器模拟人类智能的宽泛领域
2. **机器学习(ML)**：AI的分支，通过数据优化模型性能
3. **深度学习(DL)**：机器学习的重要分支，基于**多层神经网络**，实现**多级表示学习**，擅长图像、文本、语音等复杂任务

## 1.2 机器学习的基本组件
1. **数据**
    - 基本单位：**样本**，由**特征**和**标签**（监督学习）组成
    - 划分：**训练集**（拟合参数）、**测试集**（评估泛化能力）
    - 原则：独立同分布(i.i.d.)、高质量无偏见，避免垃圾进垃圾出
2. **模型**
    - 本质：可调整参数的数据映射函数
    - 深度学习：多层神经网络，自动从底层到高层提取抽象特征
3. **目标函数（损失函数）**
    - 作用：量化模型预测与真实值的差距，训练目标为**最小化损失**
    - 常用：回归用平方误差，分类用交叉熵
    - 核心问题：**过拟合**（训练集效果好，测试集效果差，泛化能力弱）
4. **优化算法**
    - 核心：**梯度下降**，沿损失减小方向迭代更新参数

## 1.3 机器学习的类型
### 1.3.1 监督学习（最常用，有标签数据）
- 核心：学习**特征→标签**的映射关系
1. 回归：输出**连续值**（房价、销量预测）
2. 分类：输出**离散类别**
    - 二分类：猫狗识别、正负情感
    - 多分类：手写数字识别
    - 多标签分类：文本多主题标注
3. 搜索与排序：网页检索、内容优先级排序
4. 推荐系统：商品、视频个性化推荐（显式/隐式反馈）
5. 序列学习：处理变长输入输出，如语音识别、机器翻译

### 1.3.2 无监督学习（无标签数据）
- 核心：挖掘数据内在结构与规律
- 典型任务：聚类、PCA降维、GAN生成数据

### 1.3.3 强化学习
- 核心：**智能体-环境-奖励**交互循环，学习最优动作策略
- 关键概念：观察、动作、奖励、策略、探索与利用
- 案例：AlphaGo、游戏AI、机器人控制

### 1.3.4 半监督/弱监督学习
- 场景：标签成本高，使用**少量标签+大量无标签数据**训练

## 1.4 深度学习发展简史
1. **早期理论基础**
    - 统计学：最小二乘法、鸢尾花数据集
    - 信息论、图灵测试、赫布学习（神经网络理论基础）
2. **蛰伏期（1995–2005）**：算力不足、数据匮乏，神经网络效果弱于核方法、决策树
3. **崛起期（2010年后）**：**大数据 + GPU算力 + 算法创新**（Dropout、注意力机制、GAN等）推动爆发

## 1.5 深度学习的核心特点
1. **端到端学习**：无需人工设计特征，模型自动提取特征
2. **多层表示学习**：逐层抽象，从底层细节到高层语义
3. **数据驱动**：数据规模决定模型上限，算法逼近上限
4. **强非线性建模**：通过激活函数提升复杂模式拟合能力
5. **跨领域通用**：一套框架适配CV、NLP、语音、医疗等领域

## 1.6 深度学习典型应用
- 计算机视觉：图像识别、自动驾驶、医学影像诊断
- 自然语言处理：机器翻译、对话系统、文本生成
- 语音领域：语音识别、语音合成
- 其他：游戏AI、推荐系统、金融风控、基因组学分析

## 1.7 核心总结
1. 机器学习四要素：**数据、模型、损失函数、优化算法**
2. 深度学习 = 多层神经网络 + 端到端训练
3. 三大学习范式：监督学习、无监督学习、强化学习
4. 崛起三要素：**大数据、算力、算法创新**
