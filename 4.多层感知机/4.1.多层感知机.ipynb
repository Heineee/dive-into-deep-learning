```python
%matplotlib inline
import torch
import sys
sys.path.append("..")
from tools.PLOT import plot
```

```python
x=torch.arange(-8.0,8.0,0.1,requires_grad=True)
y=torch.relu(x)
plot(x.detach(),y.detach(),'x','relu(x)')
```

```python
y.backward(torch.ones_like(x),retain_graph=True)
plot(x.detach(),x.grad,'x','grad of relu')
```

```python
y=torch.sigmoid(x)
plot(x.detach(),y.detach(),'x','sigmoid(x)')
```

```python
x.grad.data.zero_()
y.backward(torch.ones_like(x),retain_graph=True)
plot(x.detach(),x.grad,'x','grad of sigmoid(x)')
```

```python
y=torch.tanh(x)
plot(x.detach(),y.detach(),'x','tanh(x)')
```

```python
x.grad.data.zero_()
y.backward(torch.ones_like(x),retain_graph=True)
plot(x.detach(),x.grad,'x','grad of tanh')
```
